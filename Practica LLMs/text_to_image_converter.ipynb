```python
# 3. Importar librerías necesarias
import torch
import os
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

# Cargar el modelo
model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

# Leer el prompt desde el archivo
file_path = "/content/drive/MyDrive/LLMPractica/textfinetunned/resumen_generado.txt"
with open(file_path, "r", encoding="utf-8") as file:
    prompt = file.read()

# Generar imagen
image = pipe(prompt).images[0]

# Crear carpeta si no existe
output_dir = "/content/drive/MyDrive/LLMPractica/image_generated"
os.makedirs(output_dir, exist_ok=True)

# Guardar imagen
output_path = os.path.join(output_dir, "desc.png")
image.save(output_path)

print(f"✅ Imagen guardada en: {output_path}")
```
