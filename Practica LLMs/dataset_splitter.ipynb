```python
import os
import json
from datasets import Dataset
import pandas as pd

# 2. Cargar el JSON desde Google Drive
json_path = "/content/drive/MyDrive/LLMPractica/historical_dataset_200_visual_summaries.json"
with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 3. Convertir a Dataset de Hugging Face
dataset = Dataset.from_list(data)

# 4. Dividir en train (80%), val (10%) y test (10%)
dataset_split = dataset.train_test_split(test_size=0.2, seed=42)
train_val_split = dataset_split["train"].train_test_split(test_size=0.125, seed=42)

train_df = train_val_split["train"].to_pandas()
val_df = train_val_split["test"].to_pandas()
test_df = dataset_split["test"].to_pandas()

# 5. Guardar como CSVs en la carpeta deseada
output_path = "/content/drive/MyDrive/LLMPractica/mydatasets"
os.makedirs(output_path, exist_ok=True)

train_df.to_csv(os.path.join(output_path, "train.csv"), index=False)
val_df.to_csv(os.path.join(output_path, "val.csv"), index=False)
test_df.to_csv(os.path.join(output_path, "test.csv"), index=False)

print("✅ CSVs creados en:", output_path)
```
